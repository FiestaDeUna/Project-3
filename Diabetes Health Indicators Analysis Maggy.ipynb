# Demographics 

```python
fig, axes = plt.subplots(1, 3, figsize = (20, 8))

columns = ['Sex', 'Education', 'Income']

for i in range(len(columns)): 
    col = columns[i]
    df_diabetes[col].value_counts().plot(kind = 'pie', autopct = '%1.1f%%', ax = axes[i], title =col)
    

plt.show()

# Sex: 
# 0: 'Female' 
# 1: 'Male' 

# Income: 
# 1: Less than $10,000
# 2: Less than $15,000 ($10,000 to less than $15,000)
# 3: Less than $20,000 ($15,000 to less than $20,000)
# 4: Less than $25,000 ($20,000 to less than $25,000)
# 5: Less than $35,000 ($25,000 to less than $35,000)
# 6: Less than $50,000 ($35,000 to less than $50,000)
# 7: Less than $75,000 ($50,000 to less than $75,000)
# 8: $75,000 or more
    
# Education: 
# 1: Never attended school or only kindergarten
# 2: Grades 1 through 8 (Elementary)
# 3: Grades 9 through 11 (Some high school)
# 4: Grade 12 or GED (High school graduate)
# 5: College 1 year to 3 years (Some college or technical school)
# 6: College 4 years or more (College graduate)
```

```python
color = ['#55efc4', '#81ecec', '#74b9ff', '#a29bfe', '#00b894', '#00cec9', '#0984e3', '#6c5ce7',
         '#ffeaa7', '#fab1a0', '#ff7675', '#fd79a8', '#d63031']

# Plotting the bar chart
ax = df_diabetes['Age'].value_counts().plot(kind='bar', color=color)

# Custom labels for legend
age_groups = ['65-69', '60-64', '55-59', '70-74', '50-54', 'Over 79', '75-79',
              '45-49', '40-44', '35-39', '30-34', '25-29', '18-24']

# Create custom legend handles from empty patches and labels
legend_handles = [plt.Rectangle((0,0),1,1, color=color[i], edgecolor='black') for i in range(len(color))]

# Displaying the custom legend
plt.legend(legend_handles, age_groups)

# Title and labels
plt.title("Frequency of Age Groups")
plt.xlabel("Age Groups")
plt.ylabel("Frequency")

# Show plot
plt.show()
```

# Dataset Overview: Health Concerns

```python
color = ['#ff7675', '#00cec9']

fig, axes = plt.subplots(2, 2, figsize = (15, 15))



for i in range(len(health_issues)): 
    col = health_issues[i]
    row_index, col_index = divmod(i, 2)
    df_diabetes[col].value_counts().plot(kind = 'bar', ax = axes[row_index, col_index], title =col, color = color)
    

plt.show()
```

```python
bins = [0, 18.5, 24.9, 29.9, 40, 100]
labels = ['Underweight', 'Normal weight', 'Overweight', 'Obese', 'Morbidly obese']
df_diabetes['BMI_Category'] = pd.cut(df_diabetes['BMI'], bins=bins, labels=labels, right=False)


df_diabetes.head()
```

```python


color = ['#55efc4','#81ecec', '#74b9ff', '#a29bfe', '#00b894']

df_diabetes['BMI_Category'].value_counts().plot(kind = 'bar', color = color)

plt.title("Frequency of BMI Categories")
```

```python
df_diabetes = df_diabetes.drop(['BMI_Category'], axis=1)
 
```

# Exploratory Analysis 

```python
df_diabetes.corr(method='pearson')  
```

```python
plt.figure(figsize=(16, 6))
heatmap = sns.heatmap(df_diabetes.corr(), vmin=-1, vmax=1, annot=True, cmap="YlGnBu")
heatmap.set_title("Overall Correlation", {'fontsize':18}, pad=12)
```

```python
plt.figure(figsize=(8, 12))
heatmap = sns.heatmap(df_diabetes.corr()[['Diabetes_binary']].sort_values(by='Diabetes_binary', ascending=False), vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Correlation with Diabetes', fontdict={'fontsize':18}, pad=16)
```

```python
plt.figure(figsize=(12, 10))
df_diabetes.corr()['Diabetes_binary'].sort_values(ascending = False).plot(kind='bar')
plt.title("Correlation with Diabetes")
```
